```python
# If you want to change more column nan values use IMPUTER.
# Scikit-learn_IMPUTER_in_Python.
# Befor_IMPUTER_Must_Change_dataset_quan_dataset.
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qaul=quanQual(dataset)
```

```python
dataset[quan]
```

```python
dataset[quan].isnull().sum()
```

```python
# For more columns change nan values using IMPUTER.
# Scikit-learn_IMPUTER_in_Python.
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df = pd.DataFrame(df,columns = quan)
```
